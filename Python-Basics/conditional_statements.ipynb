## What are Conditional Statements?

Conditional statements are used to make decisions in a program.

They allow a program to execute different code based on different conditions.

Example from real life:

- If it rains → take an umbrella  
- If marks > 40 → pass  
- If balance is low → recharge  

In Python, we use:

- if  
- if-else  
- if-elif-else  


## 1. Simple IF Statement

Syntax:

if condition:
    statement


```python
age = 20

if age >= 18:
    print("You are eligible to vote")
```

## 2. IF-ELSE Statement

Used when we have two possibilities.


```python
num = 15

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

## 3. IF-ELIF-ELSE Statement

Used when there are multiple conditions.


```python
marks = 78

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

## 4. Nested IF Statements

An if statement inside another if statement is called nested if.


```python
age = 22
citizen = True

if age >= 18:
    if citizen:
        print("You can vote")
    else:
        print("You are not a citizen")
else:
    print("You are underage")
```
